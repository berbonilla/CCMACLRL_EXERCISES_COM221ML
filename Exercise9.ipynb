## 1. Train a KNN Classifier

```python
df.drop(columns="cb_person_default_on_file", inplace=True)
df.drop(columns="loan_grade", inplace=True)
df.drop(columns="id",inplace=True)
```

```python
ownership_mapping = {
    "OWN": 0,
    "RENT": 1,
    "MORTGAGE": 2,

}
loan_intent_mapping = {
    "EDUCATION": 0,
    "MEDICAL": 1,
    "VENTURE": 2,
    "PERSONAL": 3,
}
df["person_home_ownership"] = df["person_home_ownership"].map(ownership_mapping).fillna(0).astype(int)
df["loan_intent"] = df["loan_intent"].map(loan_intent_mapping).fillna(0).astype(int)
```

```python
df.head()
```

```python
df.tail()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

# Drop multiple columns by passing a list
X = df.drop(["person_home_ownership", "loan_status", "person_income"], axis=1)
y = df["loan_status"]

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, test_size=0.3)

# Initialize KNN with the best number of neighbors
KNN = KNeighborsClassifier(n_neighbors=22)  # You've found 22 to be the best value

# Fit the model
KNN.fit(X_train, y_train)

# Calculate the score
knn_score = KNN.score(X_test, y_test)
print(f"Score is {knn_score}")
```

- Perform cross validation

```python
from sklearn.model_selection import cross_val_score


cv_scores = cross_val_score(KNN, X, y, cv=5)

print("Cross-validation scores:", cv_scores)

print("Mean cross-validation score:", cv_scores.mean())
print("Standard deviation of cross-validation scores:", cv_scores.std())
```
